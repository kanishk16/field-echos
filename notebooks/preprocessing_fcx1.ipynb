# Code for preprocessing CRCNS fcx-1 dataset

```python
import numpy as np
from scipy import io
import os
import pandas as pd
```

```python
basepath = '/Users/rdgao/Documents/data/CRCNS/fcx1/'
rec_dirs = [f for f in np.sort(os.listdir(basepath)) if os.path.isdir(basepath+f)]
```

# Reorganizing behavioral data into dataframe and save out as csv

```python
for cur_rec in range(len(rec_dirs)):
    print(rec_dirs[cur_rec])
    wakesleep = io.loadmat(basepath+rec_dirs[cur_rec]+'/'+rec_dirs[cur_rec]+'_WSRestrictedIntervals.mat', squeeze_me=True)
    df_wakesleep = pd.DataFrame()
    for k in wakesleep.keys():
        # grab only the timepair matrices
        if 'TimePair' in k:
            # make sure it's not empty
            if wakesleep[k].shape[0]:
                if 'WakeSleep' in k:
                    if len(wakesleep[k].shape)==1:
                        # multiple WS periods, array of arrays
                        df_cur = pd.DataFrame([np.append(block[0,:],block[1,1]) for block in wakesleep['WakeSleepTimePairFormat']], columns=['Start', 'Inter', 'End'])
                    else:
                        # single WS period, 2x2 array
                        block = wakesleep[k]
                        df_cur = pd.DataFrame([np.append(block[0,:],block[1,1])], columns=['Start', 'Inter', 'End'])
                    df_cur.insert(0,'Label',[k[:-14]]*df_cur.shape[0])
                else:
                    # manually deal with 1-D array because pandas will assume array to be columns
                    data_block = [wakesleep[k]] if len(wakesleep[k].shape)==1 else wakesleep[k]
                    df_cur = pd.DataFrame(data_block, columns=['Start', 'End'])
                    df_cur.insert(0,'Label',[k[:-14]]*df_cur.shape[0])
                # append to overall segment
                df_wakesleep = df_wakesleep.append(df_cur, ignore_index=True)
    # rearrange labels
    df_wakesleep = df_wakesleep[['Label','Start','Inter','End']]
    # save to disk
    df_wakesleep.to_csv(basepath+rec_dirs[cur_rec]+'/'+rec_dirs[cur_rec]+'_wakesleep.csv')
```
